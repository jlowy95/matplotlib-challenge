```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
clinical_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merged = pd.merge(clinical_df,mouse_df,on="Mouse ID")

# Display the data table for preview
merged.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tv_by_drugtime = pd.DataFrame(merged.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].mean())
# Convert to DataFrame
mean_tv_by_drugtime = mean_tv_by_drugtime.reset_index()

# Preview DataFrame
mean_tv_by_drugtime.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tv_stde = merged.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem()
# Convert to DataFrame
tv_stde = pd.DataFrame(tv_stde.reset_index())
# Preview DataFrame
tv_stde.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_tv_by_drugtime = mean_tv_by_drugtime.set_index(["Drug","Timepoint"]).unstack(level=0)
mean_tv_by_drugtime.columns = mean_tv_by_drugtime.columns.droplevel(0)
mtv_ri = mean_tv_by_drugtime.reset_index()

#Perform the same for Error dataframe
tv_stde = tv_stde.set_index(["Drug","Timepoint"]).unstack(level=0)
tv_stde.columns = tv_stde.columns.droplevel(0)
# Preview that Reformatting worked
tv_stde.head()
```

```python
# Generate the Plot (with Error Bars)
mtv_plot = mtv_ri.plot(x="Timepoint", y="Capomulin",yerr=tv_stde,label="Capomulin",color="r",marker="o",lw=0.4,ms=5)
mtv_ri.plot(x="Timepoint",y="Infubinol",yerr=tv_stde,label="Infubinol",color="b",marker="^",lw=0.4,ms=5,ax=mtv_plot)
mtv_ri.plot(x="Timepoint",y="Ketapril",yerr=tv_stde,label="Ketapril",color="g",marker="s",lw=0.4,ms=5,ax=mtv_plot)
mtv_ri.plot(x="Timepoint",y="Placebo",yerr=tv_stde,label="Placebo",color="k",marker="d",lw=0.4,ms=5,ax=mtv_plot)

mtv_plot.set_xlim(-2,47)
mtv_plot.grid(axis="y")
mtv_plot.legend(["Capomulin","Infubinol","Ketapril","Placebo"])
mtv_plot.set_title("Tumor Response to Treament")
mtv_plot.set_xlabel("Time (Days)")
mtv_plot.set_ylabel("Tumor Volume (mm3)")
# Save the Figure
mtv_plot.figure.savefig("../Images/tumor_response_to_treatment.jpg")
```

```python
# Show the Figure
plt.show()
```

![Tumor Response to Treatment](../Images/treatment.png)

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
metasites_by_drugtime = merged.groupby(["Drug","Timepoint"])["Metastatic Sites"]
# Convert to DataFrame
metasites_by_drugtime = pd.DataFrame(metasites_by_drugtime.mean())
# Preview DataFrame
metasites_by_drugtime.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
ms_stde = merged.groupby(["Drug","Timepoint"])["Metastatic Sites"].sem()
# Convert to DataFrame
ms_stde = pd.DataFrame(ms_stde)
# Preview DataFrame
ms_stde.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
metasites_by_drugtime = metasites_by_drugtime.unstack(level=0)
metasites_by_drugtime.columns = metasites_by_drugtime.columns.droplevel(0)
ms_ri = metasites_by_drugtime.reset_index()

ms_stde = ms_stde.unstack(level=0)
ms_stde.columns = ms_stde.columns.droplevel(0)
# Preview that Reformatting worked
ms_stde.head()
```

```python
# Generate the Plot (with Error Bars)
ms_plot = ms_ri.plot(x="Timepoint",y="Capomulin",yerr=ms_stde,label="Capomulin",color="r",marker="o",lw=0.4,ms=5)
ms_ri.plot(x="Timepoint",y="Infubinol",yerr=ms_stde,label="Infubinol",color="b",marker="^",lw=0.4,ms=5,ax=ms_plot)
ms_ri.plot(x="Timepoint",y="Ketapril",yerr=ms_stde,label="Ketapril",color="g",marker="s",lw=0.4,ms=5,ax=ms_plot)
ms_ri.plot(x="Timepoint",y="Placebo",yerr=ms_stde,label="Placebo",color="k",marker="d",lw=0.4,ms=5,ax=ms_plot)

ms_plot.set_xlim(-2,47)
ms_plot.set_ylim(-0.25,3.8)
ms_plot.grid(axis="y")
ms_plot.set_title("Metastatic Spread During Treatment")
ms_plot.set_xlabel("Treatment Duration (Days)")
ms_plot.set_ylabel("Met. Sites")
ms_plot.legend(["Capomulin","Infubinol","Ketapril","Placebo"])
# Save the Figure
ms_plot.figure.savefig("../Images/metatstatic_spread.jpg")
# Show the Figure
ms_plot.figure.show()
```

![Metastatic Spread During Treatment](../Images/spread.png)

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count = merged.groupby(["Drug","Timepoint"])["Mouse ID"]
# Convert to DataFrame
mouse_count = pd.DataFrame(mouse_count.count())
mouse_count = mouse_count.reset_index()
mouse_count.columns = ["Drug","Timepoint","Mouse Count"]
# Preview DataFrame
mouse_count.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_count = mouse_count.set_index(["Drug","Timepoint"])
mouse_count = mouse_count.unstack(level=0)
mouse_count.columns = mouse_count.columns.droplevel(0)
mouse_count = mouse_count.reset_index()
# Preview the Data Frame
mouse_count.head()
```

```python
# Generate the Plot (Accounting for percentages)
mc_plot = mouse_count.plot(x="Timepoint",y="Capomulin",label="Capomulin",color="r",marker="o",lw=0.4,ms=5)
mouse_count.plot(x="Timepoint",y="Infubinol",label="Infubinol",color="b",marker="^",lw=0.4,ms=5,ax=mc_plot)
mouse_count.plot(x="Timepoint",y="Ketapril",label="Ketapril",color="g",marker="s",lw=0.4,ms=5,ax=mc_plot)
mouse_count.plot(x="Timepoint",y="Placebo",label="Placebo",color="k",marker="d",lw=0.4,ms=5,ax=mc_plot)

mc_plot.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=25))

mc_plot.set_title("Survival During Treatment")
mc_plot.set_xlabel("Time (Days)")
mc_plot.set_ylabel("Survival Rate")
mc_plot.grid("on")
mc_plot.set_xlim(-2,47)
mc_plot.set_ylim(8,27)
mc_plot.legend(["Capomulin","Infubinol","Ketapril","Placebo"])
# Save the Figure
mc_plot.figure.savefig("../Images/surival_percent.jpg")
# Show the Figure
plt.show()
```

![Metastatic Spread During Treatment](../Images/survival.png)
